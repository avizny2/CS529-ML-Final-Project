```python
import keras
import numpy as np
import tensorflow as tf
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    # Restrict TensorFlow to only allocate 1GB * 2 of memory on the first GPU
    try:
        tf.config.experimental.set_virtual_device_configuration(
            gpus[0],
            [tf.config.experimental.VirtualDeviceConfiguration(memory_limit=1024 * 2)])
        logical_gpus = tf.config.experimental.list_logical_devices('GPU')
        print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
    except RuntimeError as e:
        # Virtual devices must be set before GPUs have been initialized
        print(e)
```

```python
d = 100 # datapoints / feature dimensions
k = 2  # principal components
    
# Import PCA from sklearn
from sklearn.decomposition import PCA
import pandas as pd

M = np.random.normal(0, 1, (d, k))
# qr decomposition 
q,r = tf.linalg.qr(M)
#print(q)
#print(r)

B = int(d/10)

# pick different Z for different classes 
Z_1 = np.random.normal(0, 1, (k, B))
Z_2 = np.random.normal(-1, 1, (k, B))
    

for layer_i in range(0,10):
    
    #print(M)
    
    
    

    E = np.random.normal(0, 1, (d, B))
    
    X_1 = np.matmul(q, Z_1) + E
    X_2 = np.matmul(q, Z_2) + E
    
    #print(X_1) 
    #print(X_2) 

    # Instantiate PCA
    pca = PCA(n_components=k)

    # Fit PCA to features
    pcaX1 = pca.fit_transform(X_1)
    pcaX2 = pca.fit_transform(X_2)
    
    
    #print(pcaX1) 
    #print(pcaX2)
    X = np.concatenate((pcaX1, pcaX2), axis=0)
    
    # Create a new dataset from principal components 
    df = pd.DataFrame(data = X, 
                      columns = ['PC1', 'PC2'])

    target = pd.Series(np.concatenate((np.full(d, 0), np.full(d, 1)), axis=0), name='y')

    result_df = pd.concat([df, target], axis=1)
    
    #print(result_df)
    
    # Visualize Principal Components with a scatter plot
    import matplotlib.pyplot as plt 

    fig = plt.figure(figsize = (8,6))
    ax = fig.add_subplot(1,1,1) 
    ax.set_xlabel('First Principal Component ', fontsize = 15)
    ax.set_ylabel('Second Principal Component ', fontsize = 15)
    ax.set_title('Principal Component Analysis (2PCs)', fontsize = 20)

    targets = [0,1]
    colors = ['r', 'g']
    for target, color in zip(targets, colors):
        indicesToKeep = result_df.loc[result_df['y'] == target]
        #print(result_df.loc[indicesToKeep, 'PC1'])
        ax.scatter(indicesToKeep['PC1'], 
                   indicesToKeep['PC2'], 
                   c = color, 
                   s = 50)
    ax.legend(targets)
    ax.grid()
    
    
    
    
```
